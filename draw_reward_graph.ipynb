### Evaluate training result

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 300
data_regular = np.load('training_process.npy')
data_change_state = np.load('training_process_change_state.npy')
data_avoid_wall = np.load('training_process_avoid_wall.npy')
```

```python
plt.title('The regular training process.')
plt.plot(data_regular,'-',color = 'r',label="CNN-RLSTM")
plt.show()
```

```python
plt.title('The training process when I change the states.')
plt.plot(data_change_state,'-',color = 'r',label="CNN-RLSTM")
plt.show()
```

```python
plt.title('The training process when the star is always on the middle of the screen.')
plt.plot(data_avoid_wall,'-',color = 'r',label="CNN-RLSTM")
plt.show()
```
